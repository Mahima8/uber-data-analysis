```python
df['pickup_hour'] = df['Pickup Time'].dt.hour
df['pickup_day'] = df['Pickup Time'].dt.day_name()
df['pickup_month'] = df['Pickup Time'].dt.month
```

```python
df['trip_duration_min'] = (
    df['Drop Off Time'] - df['Pickup Time']
).dt.total_seconds() / 60
```

```python
import matplotlib.pyplot as plt

df['pickup_hour'].value_counts().sort_index().plot(kind='line')
plt.title("Trips by Pickup Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Trips")
plt.show()
```

```python
df['pickup_day'].value_counts().reindex(order).plot(kind='bar')
plt.title("Trips by Day of Week")
plt.show()
```

```python
plt.scatter(df['trip_distance'], df['fare_amount'], alpha=0.3)
plt.xlabel("Trip Distance")
plt.ylabel("Fare Amount")
plt.title("Fare vs Distance")
plt.show()
```
